```python
import pandas as pd
#import pandas_profiling
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
import seaborn as sns
```

```python
#data
dataset_df = pd.read_csv("data_covid19.csv")
```

```python
dataset_df
```

```python
dataset_df.info()
```

```python
dataset_df['Usia'].plot(kind='box')
```

```python
sns.distplot(dataset_df["Usia"],color="g")
```

```python
df_good = dataset_df[dataset_df["Diagnosa"] == 1]
df_bad = dataset_df[dataset_df["Diagnosa"] == 0]

fig, ax = plt.subplots(nrows=2, figsize=(12,8))
plt.subplots_adjust(hspace = 0.4, top = 0.8)

g1 = sns.distplot(df_good["Usia"], ax=ax[0], 
             color="g")
g1 = sns.distplot(df_bad["Usia"], ax=ax[1], 
             color='r')
```

```python
pd.crosstab(dataset_df["Usia"], dataset_df['Diagnosa'],margins=True)
```

```python
check = pd.crosstab(dataset_df["Usia"], dataset_df['Diagnosa'],margins=True)
```

```python
check.div(check.sum()).T.plot(kind='bar',stacked=True)
```

```python
a=[2,3,4,5,9,10]
for i in range(0,len(a)):
    plt.figure(figsize=(15,5))
    sns.boxplot(y='Usia', x=dataset_df.columns[a[i]], 
                 data=dataset_df, 
                 width=0.5,
                 palette="colorblind")
```

```python
a=[2,3,4,5,9,10]
for i in range(0,len(a)):
    plt.figure(figsize=(15,5))
    sns.violinplot(y='Usia', x=dataset_df.columns[a[i]],
                hue='Diagnosa', 
                 data=dataset_df, 
                 width=0.5,
                 palette="colorblind",split=True)
```

```python
sns.lmplot('Usia','Batuk Kering',data=dataset_df,col='Diagnosa',scatter_kws=dict(color='c'))
```
