```python
import pandas as pd
#import pandas_profiling
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
import seaborn as sns
```

```python
#data
dataset_df = pd.read_csv("data_covid19.csv")
```

```python
dataset_df['Usia'].plot(kind='box')
```

kotak panjang itu disebut whisker, boxplot biasanya dipake untuk melihat distribusi data apakah ada outlier ato tidak. warna merah tersebut dpt dikatakan bahwa pada data tsb rata usia dr hasil survey adalah ...
apabila ada titik kecil dilur kotak maka titik kecil tsb menunjukan adanya outlier/pencilan data. outlier tidak perlu dihapus krn hal tsb dapat mengetahui adanya anomali pada data.

```python
dataset_df['Usia'].plot(kind='hist')
```

```python
sns.distplot(dataset_df["Usia"],color="g")
```

visualisasi ini jg untuk melihat distribusi frekuensi data pada kolom usia. Fungsi ini akan memvisualisasikan histogram yang sesuai dengan estimasi kepadatan kernel/Kernel Density Estimation (KDE) dari data

```python
df_good = dataset_df[dataset_df["Diagnosa"] == 1]
df_bad = dataset_df[dataset_df["Diagnosa"] == 0]

fig, ax = plt.subplots(nrows=2, figsize=(12,8))
plt.subplots_adjust(hspace = 0.4, top = 0.8)

g1 = sns.distplot(df_good["Usia"], ax=ax[0], 
             color="g")
g1 = sns.distplot(df_bad["Usia"], ax=ax[1], 
             color='r')
```
